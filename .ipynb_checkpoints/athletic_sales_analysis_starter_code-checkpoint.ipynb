```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
df_2020.head()
```

```python
# Display the 2021 sales DataFrame
df_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
df_2021.shape
```

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df_all = pd.concat([df_2020,df_2021], axis=0, join='inner', ignore_index=True)
df_all.head(15)
```

```python
# Check if any values are null.
df_all.count()
```

```python
# Check the data type of each column
df_all.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
df_all['invoice_date'] = pd.to_datetime(df_all['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
df_all.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# Show the top 5 results.

df_all.groupby(['region','state','city'])[['units_sold']].sum().sort_values(by='units_sold', ascending=False).rename(columns={'units_sold':'Total_Products_Sold'}).head()                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Rename the "units_sold" column to "Total_Products_Sold"
# Show the top 5 results.

df_all.pivot_table(index=['region','state','city'], values='units_sold',aggfunc='sum').sort_values(by='units_sold', ascending=False).rename(columns={'units_sold':'Total_Products_Sold'}).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

df_all.groupby(['region','state','city'])[['total_sales']].sum().sort_values(by='total_sales', ascending=False).rename(columns={'total_sales':'Total Sales'}).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

df_all.pivot_table(index=['region','state','city'], values='total_sales',aggfunc='sum').sort_values(by='total_sales', ascending=False).rename(columns={'total_sales':'Total Sales'}).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

df_all.groupby(['retailer','region','state','city'])[['total_sales']].sum().sort_values(by='total_sales', ascending=False).rename(columns={'total_sales':'Total Sales'}).head()

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the top 5 results.

df_all.pivot_table(index=['retailer','region','state','city'], values='total_sales',aggfunc='sum').sort_values(by='total_sales', ascending=False).rename(columns={'total_sales':'Total Sales'}).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_data = df_all.loc[df_all['product'] == "Women's Athletic Footwear"].copy()
womens_data['product'].unique()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.

womens_data.groupby(['retailer','region','state','city'])[['units_sold']].sum().sort_values(by='units_sold', ascending=False).rename(columns={'units_sold':'Womens_Footwear_Units_Sold'}).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# Show the top 5 results.

womens_data.pivot_table(index=['retailer','region','state','city'], values='units_sold',aggfunc='sum').sort_values(by='units_sold', ascending=False).rename(columns={'units_sold':'Womens_Footwear_Units_Sold Sales'}).head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
womens_data.pivot_table(index='invoice_date',values='total_sales', aggfunc='sum')
```

```python
# Create a pivot tabe with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
# Show the table.

womens_pivot_table = womens_data.pivot_table(index='invoice_date',values='total_sales', aggfunc='sum').rename(columns={'total_sales':'Total Sales'})
womens_pivot_table

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in ascending order on "Total Sales".

womens_pivot_table.resample('D').sum().sort_values(by='Total Sales')
```

```python
#Nico comment: The instructions in bootcampspot say in descending order and the instructions in the above cell say in ascending order
# I'll do it in both so that I don't get points deducted.  Ascending is above, and this one is descending
# Also, the instructions in the cell above don't mention to display the top 10 so I didn't do a head() call above, but since the 
# instructions in bootcampspot say to display the top 10 days for the descending order one, i'll do that call here.

womens_pivot_table.resample('D').sum().sort_values(by='Total Sales', ascending=False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in ascending order on "Total Sales".

womens_pivot_table.resample('W').sum().sort_values(by='Total Sales')

```
